```python
import pandas as pd 
import seaborn as sns
```

```python
#Importing Data 

mammal_df = pd.read_csv("mammals.csv")
```

```python
mammal_df.shape
```

```python
#drops rows with null values 
mammal_df.dropna(axis = 0, how = 'any', inplace = True) 
```

```python
#dropping columns 
mammal_df.drop('predation', axis= 1, inplace = True)
mammal_df.drop('exposure', axis= 1, inplace = True)
mammal_df.drop('danger', axis= 1, inplace = True)
mammal_df.shape
```

```python
sns.set_theme(style = 'darkgrid')
# Line graph for total sleep and gestation
sns.lineplot(x="total_sleep", y="gestation",
             data=mammal_df).set(title = 'Gestation Time and Total Sleep' , xlabel = 'Total Sleep (hours)' , ylabel = 'Gestation (days)')
```

```python
#plot for total sleep and lifespan 
sns.set_theme(style = 'darkgrid')
sns.relplot(data = mammal_df,x="total_sleep", y="life_span",).set(title = 'Total Sleep and Life Span' , xlabel = 'Total Sleep (hours)', 
            ylabel = 'Life Span (Years)')
```
